```python
def apply_operation(row, col_name,eVperAtom, dt):
    a = row[col_name]
    if row.name == "Antisitepair":
        return a
    elif row.name == "C_Ta":
        return a + dt.loc["V_C", col_name] - dt.loc["V_Ta", col_name]
    elif row.name == "divacancy":
        return a + 2 * dt.loc["nodefect", eVperAtom]
    elif row.name == "Ta_C":
        return a - dt.loc["V_C", col_name] + dt.loc["V_Ta", col_name]
    elif row.name == "V_TaV_2C_Line":
        return 0.5*(2*a + dt.loc["V_Ta",col_name] - dt.loc["V_C",col_name] + 6*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_2C_Tri":
        return 0.5*(2*a + dt.loc["V_Ta",col_name] - dt.loc["V_C",col_name] + 6*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_3C_InPlane":
        return (a + dt.loc["V_Ta",col_name] - dt.loc["V_C",col_name] + 4*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_3C_OutPlane":
        return (a + dt.loc["V_Ta",col_name] - dt.loc["V_C",col_name] + 4*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_4C_InPlane":
        return 0.5*(2*a + 3*dt.loc["V_Ta",col_name] - 3*dt.loc["V_C",col_name] + 10*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_4C_OutPlane":
        return 0.5*(2*a + 3*dt.loc["V_Ta",col_name] - 3*dt.loc["V_C",col_name] + 10*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_5C":
        return (a + 2*dt.loc["V_Ta",col_name] - 2*dt.loc["V_C",col_name] + 6*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_6C":
        return 0.5*(2*a + 5*dt.loc["V_Ta",col_name] - 5*dt.loc["V_C",col_name] + 14*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_Ta":
        return 0.5*(a + dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_C":
        return 0.5*(a + dt.loc["V_Ta",col_name] + 2*dt.loc["nodefect",eVperAtom])
    else:
        return 0

dt['EFE-DFT-Stoich'] = dt.apply(apply_operation, args=("DFT-Raw(eV)","DFT(eV/Atom)", dt), axis=1)
dt['EFE-PINN-Stoich'] = dt.apply(apply_operation, args=("PINN-Raw(eV)","PINN(eV/Atom)", dt), axis=1)
dt
```

```python
fig,ax = plt.subplots()
ax.plot(dt["EFE-DFT-Stoich"],"rs")
ax.plot(dt["EFE-PINN-Stoich"],"bo")
ax.set_ylabel("Effective Formation Energy (eV)")
ax.set_title("Effective Formation Energy Stoichiometry (eV)")
ax.legend(["DFT", "PINN"])
ax.set_xticks(range(len(dt)))
ax.set_xticklabels(dt.index, rotation=90)
indices = [12,13,14]
for i in indices:
    ax.get_xticklabels()[i].set_color("green")
```

```python
def apply_B_defi_operation(row, col_name,eVperAtom, dt):
    a = row[col_name]
    if row.name == "Antisitepair":
        return a
    elif row.name == "C_Ta":
        return a + 2*dt.loc["V_C", col_name] + 2*dt.loc["nodefect", eVperAtom]
    elif row.name == "divacancy":
        return a + 2 * dt.loc["nodefect", eVperAtom]
    elif row.name == "Ta_C":
        return a - 2*dt.loc["V_C", col_name] - 2*dt.loc["nodefect", eVperAtom]
    elif row.name == "V_TaV_2C_Line":
        return (a - dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_2C_Tri":
        return (a - dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_3C_InPlane":
        return (a  - 2*dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_3C_OutPlane":
        return (a  - 2*dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_4C_InPlane":
        return (a - 3*dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_4C_OutPlane":
         return (a - 3*dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_5C":
        return (a - 4*dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_TaV_6C":
        return (a - 5*dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_Ta":
        return (a + dt.loc["V_C",col_name] + 2*dt.loc["nodefect",eVperAtom])
    elif row.name == "V_C":
        return 0
    else:
        return 0
   
dt['EFE-DFT-B-defi'] = dt.apply(apply_B_defi_operation, args=("DFT-Raw(eV)","DFT(eV/Atom)", dt), axis=1)
dt['EFE-PINN-B-defi'] = dt.apply(apply_B_defi_operation, args=("PINN-Raw(eV)","PINN(eV/Atom)", dt), axis=1)
dt
```

```python
fig,ax = plt.subplots()
ax.plot(dt["EFE-DFT-B-defi"],"rs")
ax.plot(dt["EFE-PINN-B-defi"],"bo")
ax.set_ylabel("Effective Formation Energy (eV)")
ax.set_title("Effective Formation Energy B-deficient (eV)")
ax.legend(["DFT", "PINN"])
ax.set_xticks(range(len(dt)))
ax.set_xticklabels(dt.index, rotation=90)
indices = [12,13,14]
for i in indices:
    ax.get_xticklabels()[i].set_color("green")
```

```python
fig,ax = plt.subplots()
ax.plot(dt["EFE-DFT-B-defi"] - dt["EFE-PINN-B-defi"],"ko")
#ax.plot(dt["EFE-PINN-Stoich"],"bo")
ax.set_ylabel("Dev. EFE (eV)")
ax.set_title("Deviation in B-deficient EFE(eV) i.e. DFT - PINN")
#ax.legend(["DFT", "PINN"])
ax.set_xticks(range(len(dt)))
ax.set_xticklabels(dt.index, rotation=90)
ax.set_ylim([-0.1, 0.11])
indices = [12,13,14]
for i in indices:
    ax.get_xticklabels()[i].set_color("green")
```
